```python
# %matplotlib
# Import our dependencies
import pandas as pd
import matplotlib as plt
from sklearn.datasets import make_blobs
import sklearn as skl
# import tensorflow as tfb
```

```python
fips_df = pd.read_csv('count_fips.csv')
fips_df.head()
```

```python
county_CoC = pd.read_csv('fips_CoC.csv')

county_CoC.head()
```

```python
county_CoC2_df = pd.merge(fips_df, county_CoC, on=["FIPS", "FIPS"])
county_CoC2_df.head()
```

```python
list(county_CoC2_df.columns)
```

```python
CoC_FIPS_df = county_CoC2_df.drop(["Name", "State"], axis=1)
CoC_FIPS_df.head()
```

```python
population_county_df = pd.read_csv('co-est2019-annres.csv')

population_county_df.head()
```

```python
population_county_df = population_county_df.rename(columns = {'county':'countystate'})
population_county_df.head()
```

```python
county_CoC3_df = CoC_FIPS_df.join(population_county_df)
county_CoC3_df.head()
```

```python
CoC_FIPS_df = county_CoC3_df.drop(["countystate"], axis=1)
CoC_FIPS_df.head()
```

```python
poverty_df = pd.read_csv("PovertyEstimates_cleaned.csv")
poverty_df.head()
```

```python
county_CoC4_df = pd.merge(CoC_FIPS_df, poverty_df, left_on=["FIPS"], right_on = ["FIPStxt"], how = 'left')
county_CoC4_df.head()
```

```python
CoC_FIPS_df = county_CoC4_df.drop(["Poverty 0-4", "Pct Poverty 0-4"], axis=1)
CoC_FIPS_df.head()
```
